### i.	Concatenate two data frames using the following two dictionaries

```python
GhanaCars = {'Company': ['KANTAKA', 'VW', 'BMV', 'NEOPLAN'],
             'Price':[23845, 171995, 135925 , 71400]}#Dictionary for Ghana Cars
JapaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 
                'Price': [29995, 23600, 61500 , 58900]}#Dictionary for Japan Cars
Ghcar =pd.DataFrame.from_dict(GhanaCars)#Creating DataFrame from the Dictionary
Jacar =pd.DataFrame.from_dict(JapaneseCars)#Creating DataFrame from the Dictionary

carsdf = pd.concat([Ghcar,Jacar], keys=["Ghana",'Japan'])#concatenating and keying
carsdf
```
